###Number of connections By Period

```python
# prompt: install  plotly

!pip install plotly
import pandas as pd
import numpy as np
import plotly.express as px
import datetime
```

```python
connections
```

```python
from datetime import datetime

def change_date(date_str):
    return datetime.strptime(date_str, '%Y-%m-%d').strftime('%Y-%m-%d') # Change the format string to match the date format in your data

connections['Connected On'] = connections['Connected On'].apply(change_date)

connections
```

```python
graph = px.scatter(connections, x = 'Full Name', y = 'Connected On')
graph.show()
```

```python
connections.groupby(by = 'Connected On').count()
```

```python
graph = px.line(connections.groupby(by = 'Connected On').count(),title= 'New Connections per date')
graph.show()
```

```python
# prompt: create two new columns in dataframe to store the month and year when connections were made between users. Then generate graph of new connections per month and another graph of new connections per year.

import pandas as pd
import plotly.express as px

# Create two new columns to store the month and year
connections['Month'] = pd.to_datetime(connections['Connected On']).dt.month
connections['Year'] = pd.to_datetime(connections['Connected On']).dt.year

# Generate a graph of new connections per month
graph = px.line(connections.groupby('Month').size(),
                 title='New Connections per Month')
graph.show()

# Generate a graph of new connections per year
graph = px.line(connections.groupby('Year').size(),
                 title='New Connections per Year')
graph.show()
```

```python
def set_month(date):
  return datetime.strptime(date, '%Y-%m-%d').strftime('%m')
```

```python
connections['Connected Month'] = connections['Connected On'].apply(set_month)
connections
```

```python
connections.groupby(by = 'Connected Month').count()
```

```python
graph = px.line(connections.groupby(by = 'Connected Month').count(), title= 'New Connections by Month')
graph.show()
```

```python
def set_year(date):
  return datetime.strptime(date, '%Y-%m-%d').strftime('%Y')
```

```python
connections['Connected Year']= connections['Connected On'].apply(set_year)
connections
```

```python
connections.groupby(by = 'Connected Year').count()
```

```python
graph = px.line(connections.groupby(by = 'Connected Year').count(), title = 'New connections By Year')
graph.show()
```
