```python
# prompt: create two new columns in dataframe to store the month and year when connections were made between users. Then generate graph of new connections per month and another graph of new connections per year.

import pandas as pd
import plotly.express as px

# Create two new columns to store the month and year
connections['Month'] = pd.to_datetime(connections['Connected On']).dt.month
connections['Year'] = pd.to_datetime(connections['Connected On']).dt.year

# Generate a graph of new connections per month
graph = px.line(connections.groupby('Month').size(),
                 title='New Connections per Month')
graph.show()

# Generate a graph of new connections per year
graph = px.line(connections.groupby('Year').size(),
                 title='New Connections per Year')
graph.show()
```

```python
from datetime import datetime  # Import the datetime class

def set_month(date):
  return datetime.strptime(date, '%d-%b-%y').strftime('%m') # Change format string to match date

connections['Connected Month'] = connections['Connected On'].apply(set_month)
connections
```

```python
connections.groupby(by = 'Connected Month').count()
```

```python
graph = px.line(connections.groupby(by = 'Connected Month').count(), title= 'New Connections by Month')
graph.show()
```

```python
def set_year(date):
  return datetime.strptime(date, '%d-%b-%y').strftime('%Y') # Change format string to match date
```

```python
connections['Connected Year']= connections['Connected On'].apply(set_year)
connections
```

#**N-Gram Similarity**

##### way of expressing every possible consecutive sequence of n tokens of a text

##### We compute the 'bigrams' for all tokens of two strings and score the similarity by counting the no. of common bigrams

```python
bigrams_position1 = nltk.bigrams('business intelligence analyst')
```

```python
bigrams_position1 = list(nltk.bigrams('business intellignce analyst'))
bigrams_position1
```

```python
'business intelligence analyst'.split()
bigrams_position1 = list(nltk.bigrams('business intellignce analyst'.split()))
bigrams_position1
```

```python
bigrams_position1 = list(nltk.bigrams('business intelligence analyst'.split(),pad_right= True, pad_left = True))
bigrams_position1
```

```python
bigrams_position2 = list(nltk.bigrams('data scientist'.split(), pad_right= True, pad_left= True))
bigrams_position2
```

```python
bigrams_position3 = list(nltk.bigrams('data analyst'.split(), pad_right= True, pad_left= True))
bigrams_position3
```

```python
set(['A','B','C']).intersection(set(['C','A','F']))
```

```python
# Position 1 * Position 2
set(bigrams_position1).intersection(set(bigrams_position2))
```

```python
len(set(bigrams_position1).intersection(set(bigrams_position2)))
```

```python
len(set(bigrams_position1).intersection(set(bigrams_position3)))
```

```python
len(set(bigrams_position2).intersection(set(bigrams_position3)))
```

#** Jaccard distance**

##### No. of items in common divided by the total no. of distinct items in two sets

##### Use n - grams

```python
position1 = 'business intelligence engineer'.split()
position2 = 'data scientist'.split()
position3 = 'data analyst'.split()
```

```python
position1,position2,position3
```

```python
#position2 * position3
```

```python
intersection = set(position2).intersection(set(position3))
intersection
```

```python
union = set(position2).union(set(position3))
union
```

```python
len(intersection), len(union)
```

```python
(len(union) - len(intersection))/ len(union)
```

lower the value more the similar strings are

```python
from nltk.metrics.distance import jaccard_distance
```

```python
# Position 1 * Position 2
jaccard_distance(set(position1), set(position2))
```

```python
jaccard_distance(set(position1), set(position3))
```

```python
jaccard_distance(set(position2), set(position3))
```
